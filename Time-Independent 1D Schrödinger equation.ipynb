# The Schrödinger Equation

The Time-Dependent Schrödinger Equation (TDSE) was first published by [E. Schrödinger (1926)](https://doi.org/10.1103/PhysRev.28.1049). In its operator form, the TDSE can be written as:

$$\hat{H} |\Psi(t)\rangle = i \hbar \frac{\partial}{\partial t} |\Psi(t)\rangle.$$

Here, $\hat{H}$ is the Hamiltonian operator associated with the particular system that we are studying. For example, for a particle subject to some potential $V(x,t)$, the Hamiltonian is given by $\hat{H} = \frac{\hat{p}^2}{2m} + \hat{V}$, where $\hat{p}$ is the momentum operator and $\hat{V}$ is the potential operator. These operators act on the infinite-dimensional vector space formed by the solutions to the Schrödinger equation $|\Psi(t)\rangle$, which is known as the Hilbert space of solutions $\mathcal{H}$. In quantum mechanics, all physical observables are represented by Hermitian operators $\hat{A}$ acting on $\mathcal{H}$. From linear algebra, we know that the set of eigenvectors of a Hermitian operator form an orthonormal basis. If the eigenvectors of a given operator over $\mathcal{H}$ are known, any vector $|\Psi(t)\rangle \in \mathcal{H}$ can be written as a linear comination of them, and thus, the expected value of the associated observable can be easily obtained using this decomposition when computing $ \langle a (t) \rangle = \langle \Psi(t) | \hat{A} | \Psi(t) \rangle$. The TDSE, as we have expressed it above, is written in abstract mathematical form. However, for a system consisting of a single particle with momentum and position operators $\hat{\boldsymbol{p}}, \hat{\boldsymbol{r}}$, we can express it in its spatial representation as follows:

$$ \left[- \frac{1}{2m} \boldsymbol{\nabla}^2_\boldsymbol{r} + V(\boldsymbol{r},t) \right] \Psi(\boldsymbol{r},t) = i \hbar \frac{\partial}{\partial t}  \Psi(\boldsymbol{r},t).$$

The solutions $\Psi(\boldsymbol{r},t)=\langle \boldsymbol{r} | \Psi (t) \rangle$, where $ |\boldsymbol{r}\rangle$ is an eigenstate of the position operator with eigenvalue $\boldsymbol{r}$, are known as wave functions. The wave function of a system, which takes a complex value for each position $\boldsymbol{r}$ and time $t$, has no physical meaning in itself. Nevertheless, its modulus square $P(\boldsymbol{r},t)=|\Psi(\boldsymbol{r},t)|^2$ corresponds to the probability density function for a measurement of the particle's position at time $t$. The probability density $P(\boldsymbol{r},t)$ has units volume$^{-1}$, and it can be used to determine the likelihood of finding the particle in a given region of space $V$ at a time $t$ by calculating $P_V(t) = \int_V |\Psi(\boldsymbol{r},t)|^2 d^3 r$ . For wave functions that represent physical systems, this imposes the condition $\langle \Psi(t) | \Psi(t) \rangle = \int_{-\infty}^{\infty} |\Psi(\boldsymbol{r},t)|^2 d^3 r = 1$, which can be enforced by means of renormalization. In fact, to verify that a solution to the Schrödinger equation describes a physical system the following conditions must be upheld:

- The wave function and its spatial derivatives must be continuous.
- The wave function must take finite values for each position at all times.
- The spatial integral of the wave fuction's square $\int d^3 r\, \Psi \Psi^*$ must converge. 
- The wave function must fulfill the boundary conditions characteristic to the problem at hand.

## Time Independent Schrödinger Equation (TISE)

Considering a time independent Hamiltonian and using the separation of variables method on the spatial representation of the TDSE, one may write a time independent version of the Schrödinger equation as follows:

$$\hat{H} |\psi\rangle = E|\psi \rangle,$$

where $|\psi\rangle$ is the 'stationary' state of the system, and the separation constant $E$ is a real number corresponding to the energy of the state. This is known as an eigenvalue problem, where one must find all the possible eigenvalues $E_n$ of the operator $\hat{H}$, along with the corresponding eigenvectors $|\psi_n\rangle$. In this case, the time evolution of a stationary state is given by $|\psi_n (t)\rangle = e^{-iE_nt/\hbar} |\psi_n\rangle$. There are infinitely many eigenvalues $E_n$ for which the TISE holds, and, in general, the index (or indices) $n$ may be discrete or continuous. It can be shown that for particles whose position is constrained to a defined volume (localized), the energy spectrum is discrete; while for unlocalized particles the energy spectrum is continuous. A general state $|\Psi (t) \rangle = \sum_n c_n(t) |\psi_n\rangle$, where $c_n(t)=\langle \psi_n | \Psi (t)\rangle$, is an element of the Hilbert space $\mathcal{H}$ expressed in terms of the basis of eigenvectors of $\hat{H}$, and therefore, a solution to the original TDSE.

## Numerical Solution for the 1D TISE

Consider a time-independent one-dimensional Hamiltonian $\hat{H}=-\frac{\hbar}{2m}\frac{d^2}{dx^2} + \hat{V}(x)$ for a single localized particle with boundary conditions $\Psi(-L/2,t)=\Psi(L/2,t)=0 \,\, \forall t$. Then, a simple code using the finite difference method can provide solutions for any given potential defined in the interval $[-L/2,L/2]$. We begin by expressing the TISE as a discrete equation for which we consider a grid of $N=L/\Delta x$ points between $-L/2$ and $L/2$. Using the finite difference prescription for the second derivative we can write one equation for each lattice point $x_i$:

$$-\frac{\hbar}{2m}\left[ \frac{\psi(x_{i+1})-2\psi(x_i)+\psi(x_{i-1})}{\Delta x ^2} \right] + V(x_i) \psi(x_i)=E \psi(x_i).$$

The equations for points $x_0$ and $x_N$ are given by the boundary conditions, leaving a system of $N-1$ equations. Writing the system in matrix form we have:

$$\begin{bmatrix} \frac{\hbar}{m \Delta x^2}+V(x_i) & -\frac{\hbar}{2m \Delta x^2} & 0 & 0 & \dots & 0 \\ -\frac{\hbar}{2m \Delta x^2} & \frac{\hbar}{m \Delta x^2}+V(x_i) & -\frac{\hbar}{2m \Delta x^2} & 0 & \dots & 0 \\  \cdot & & & & & \cdot \\ \cdot &  &  &  &  & \cdot  \\ \cdot &  &  &  &  & \cdot \\ 0 & \dots & 0 & -\frac{\hbar}{2m \Delta x^2} & \frac{\hbar}{m \Delta x^2}+V(x_i) & -\frac{\hbar}{2m \Delta x^2}  \\ 0 & \dots & 0 & 0 & -\frac{\hbar}{2m \Delta x^2} & \frac{\hbar}{m \Delta x^2}+V(x_i) \end{bmatrix} \begin{bmatrix} \Psi (x_1) \\ \Psi (x_2) \\   \\ \cdot \\ \cdot \\ \cdot \\   \\  \Psi (x_{N-1}) \end{bmatrix} = E  \begin{bmatrix} \Psi (x_1) \\ \Psi (x_2) \\   \\ \cdot \\ \cdot \\ \cdot \\   \\  \Psi (x_{N-1}) \end{bmatrix} $$

Again, we have an eigenvalue problem. The solution to the system of equations will be determined if we manage to diagonalize the matrix that represents it, where the diagonal entries correspond to the eigenvalues. The system is represented by a symmetric tridiagonal Toeplitz matrix, which is easy to diagonalize. Tridiagonal Toeplitz matrices have a simple [analytical expression](https://en.wikipedia.org/wiki/Tridiagonal_matrix#cite_note-14) for their eigenvalues. Nevertheless, here we use the scipy.linalg function [eigh_tridiagonal](https://docs.scipy.org/doc/scipy/reference/generated/scipy.linalg.eigh_tridiagonal.html) which solves the eigenvalue problem for real symmetric tridiagonal matrices.

Note that this approach approximates continuous space as an N-point lattice on a defined interval $[-L/2,L/2]$. Therefore, the generally infinite energy spectrum is replaced by a finite number of discrete values $E_n$. As a result, the associated eigenfunctions are only approximately correct. This approximation improves with increasing number of lattice points $N$, and therefore, by reducing the separation $\Delta x$.

In the following cells, the well-known square well potential problem is solved using this method. If the reader wishes to learn more about the topics discussed here, I recommend the following sources:

```python
#Here I import the modules and functions we will use
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh_tridiagonal
from scipy.integrate import simpson
import time
```

## The potential well
Consider a particle that experiences a square well potential of width $a$:

$$V(x) = \biggl\{ \begin{matrix} V(x)=0 & \text{if} & x\notin [-a/2,a/2] \\ V(x)=-V_0 & \text{if} & x \in [-a/2,a/2], \end{matrix} $$

where the potential constant $V_0$ is a positive real number. In the following cell, we make a specific choice of parameters and plot the potential function we will be considering.

```python
bohr = 5.29177210903e-11 #m (bohr radius)
ion = 2.18e-18 #J (hydrogen ionization energy)
elec = 9.1093837e-31 # kg (mass of an electron)
a=1e9*bohr #m (bohr radius)
V0=1e9*ion #J (hydrogen ionization energy)
def V(x):
    if x<=-a/2:
        val=0
    elif x>=a/2:
        val=0
    else:
        val=-V0
    return val
margin=a/4
x0=-a/2-margin
xf=a/2+margin
n=2000
dom=np.linspace(x0,xf,n)
pot=[V(i) for i in dom]
fig,ax =plt.subplots(1)
plt.plot(dom,pot)
plt.axhline(y=0, color='dimgray', linestyle='--',linewidth=0.7)
plt.axvline(x=0, color='dimgray', linestyle='--',linewidth=0.7)
#ax.set_aspect('equal')
ax.set_title("Potential function", fontsize=15)
ax.set_ylabel("$V(x)$ [J]")
ax.set_xlabel("$x \, [m]$")
plt.show()
```

For this type of potential, we may consider two distinct cases:
1. When the particle has negative energy - bounded solutions.
2. When the particle has positive energy - delocalized solutions.

Situation 1 satisfies the requirement that the particle be localized in space. Trying to solve analytically for negative energies leads to a transcendental equation that can be solved with numerical methods. With this equation, it can be shown that the finite square well always has at least one bounded state, and that to have more, the condition $a^2V_0 > \pi^2 \hbar^2/2m$ must hold. Additionally, we know that as $V_0\longrightarrow\infty$ the bounded eigenfunctions will approximate the known infinite well solutions. Here, instead of solving the aforementioned equation (which one can find in any introductory quantum mechanics text), we directly apply the numerical method presented above. The boundary condition for this problem is $\Psi(-\infty)=\Psi(\infty)=0$. Therefore, our method should be a good approximation if $L>>a$. 

In situation 2, on the other hand, the potential doesn't capture the particle. Instead, it scatters it with a given transmission/reflection rate. In this case, the wave function at $\pm\infty$ will be that of a free particle, and our method won't be adequate to solve the problem.

In the following cells, we solve the eigenvalue problem for the negative energies using $L=10a$ and $\Delta x= a/10^6$.

```python
hbar = 1.05457182e-34 # m^2 kg / s
pmass = 1e9*elec # kg
L=10*a
deltx=a/1e6
lattice = np.arange(-L/2,L/2,deltx) 
potential = np.array([V(x) for x in lattice])
N = len(lattice)
print("N = ", N)
print("deltx = ", deltx)
```

Now we have everything we need to solve the system. The function eigh_tridiagonal takes as input the diagonal of lenght $N-1$ and the off-diagonal of lenght $N-2$, and outputs the eigenvalues and the normalized eigenvectors$^*$.

$*$This is normalized in the sense of the dot product, not the product $\langle \psi_n | \psi_n \rangle = \int |\psi_n(x)|^2 dx = \int \psi_n(x)\psi^*_n(x) dx$. Therefore, we must normalize using the correct definition with the simpson function.

```python
d = potential[1:-1] + hbar / (pmass * deltx**2)
e = - hbar / (2 * pmass * deltx**2) * np.ones(len(d)-1)
```

```python
#Numerical solution
initial = time.time()
bound_val, bound_vec = eigh_tridiagonal(d, e, select='v', select_range=(-V0,0))
print(len(bound_val))
#Normalize solutions
nbound_vec = np.zeros([len(bound_val),len(bound_vec.T[0])])
for i in range(len(bound_val)):
    norm = simpson(y = np.abs(bound_vec.T[i])**2, x = lattice[1:-1])
    nbound_vec[i][:] = bound_vec.T[i]/np.sqrt(norm)
    new_norm = simpson(y = np.abs(nbound_vec[i])**2, x = lattice[1:-1])
    print(norm)
    print(new_norm)
    print()
final = time.time()
print('Computation time = ', (final-initial), " sec")
```

```python
#plot the energy eigenvalues
fig, ax1= plt.subplots(1,figsize=[5,5])
ax1.bar(range(len(bound_val)),bound_val)
ax1.set_title("Bounded state energies", fontsize=15)
ax1.set_xlabel("n", fontsize=13)
ax1.set_ylabel("$E_n$ [J]", fontsize=13)
ax1.axhline(y=-V0, color='gray', linestyle='--')
#ax1.ticklabel_format(style='sci', axis='y', scilimits=(-4,-4))
ax1.tick_params(labelsize=12)
ax1.text(3.1,-V0+5e-11,"$-V_0$", color='gray', fontsize=13)
plt.tight_layout()
plt.show()
```

```python
#Let's plot the bounded eigenfunctions
fig= plt.figure(figsize=[8,len(bound_val)*5])
for i in range(len(bound_val)):
    ax = fig.add_subplot(900+10+i+1)
    graph,=ax.plot(lattice[3700001:-3700001], nbound_vec[i][3700000:-3700000], color='indigo', label="Wave function")
    ax.set_ylabel(r"$\psi_n(x)$")
    ax.set_xlabel("$x$ [m]")
    ax.yaxis.label.set_color('indigo')
    ax.tick_params(axis='y', colors='indigo')
    ax.set_ylim(-7,7)
    ax.text(lattice[3700001],5,"n = "+str(i+1), fontsize=10)
    ax.axhline(y=0,color='indigo', linestyle='--',linewidth=0.7)
    ax.axvline(x=0,color='dimgray', linestyle='--',linewidth=0.7)
    ax.axvline(x=-a/2,color='black', linestyle='--',linewidth=0.7)
    ax.axvline(x=a/2,color='black', linestyle='--',linewidth=0.7)
    ax1=ax.twinx()
    graph1,=ax1.plot(lattice[3700001:-3700001], np.abs(nbound_vec[i][3700000:-3700000])**2, color='orange', label="Probability distribution")
    ax1.axhline(y=0,color='orange', linestyle='--',linewidth=0.7)
    ax1.set_ylabel(r"$|\psi_n(x)|^2$ [1/m^3]")
    ax1.yaxis.label.set_color('orange')
    ax1.tick_params(axis='y', colors='orange')
    ax1.set_ylim(-5,35)
    plt.legend(handles=[graph,graph1], loc="upper right", fontsize=9)
plt.tight_layout()
plt.show()
```

In the figure above, we see that the solutions for odd $n$ correspond to even wave functions, while even $n$ corresponds to odd wavefunctions, as is the case for the solutions of the infinite square well. We also see that, unlike the classical or infinite square well solutions, there is a non-zero probability of finding a particle with negative energy outside the well, a hallmark of quantum mechanics.
